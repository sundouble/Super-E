```
#read .txt file with Chinese characters
import pandas as pd
import numpy as np

file_object=open("slist.txt", encoding="gbk")
try:
    file_content=file_object.read()
finally:
    file_object.close()


a=[]
result=file_content.split('\t')
for i in result:
    a.append(i)
#将里面数字变成int
a=list(a)
#将其变成array
#a=np.array(a)
#重组数据变成你想要的矩阵, 60*10
#MAT=a.reshape((2,300))
#输出为excel
#pd.Dataframe(MAT).to_excel("D:/处理过的txt.xlsx")
print(a)
#data_df=yf.download("000425.sz",interval="15m",start="2020-04-01")
#print(data_df)
for i in a:
  print(i)
```

```
file_object1=open("nlist.txt", encoding="gbk")
try:
    file_content1=file_object1.read()
finally:
    file_object1.close()


b=[]
result1=file_content1.split('\t')
for i in result1:
    b.append(i)
#将里面数字变成int
b=list(b)
#将其变成array
#a=np.array(a)
#重组数据变成你想要的矩阵, 60*10
#MAT=a.reshape((2,300))
#输出为excel
#pd.Dataframe(MAT).to_excel("D:/处理过的txt.xlsx")
print(b)
```
